```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
cnt=0
index=0
organ_name=[]
organ_content=[]
```

```python
def data_generation(cnt,url):
  content=''
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  for cont in ( (soup.find_all("span", class_='toc_subsection toc_links')) ):    
    content=content+' '+str(cont.text)
  for cont1 in (soup.find_all("div", class_='toc_subsection toc_links')):
    for cont in cont1.find_all('a'):
      content=content+'\n'+str(cont.text)
  for cont2 in (soup.find_all("div", class_='toc_links')):
    for cont in cont2.find_all('a'):
      content=content+'\n'+str(cont.text)
  content = re.sub(r"\n",r", ",content)
  organ_content.insert(cnt, content)
```

```python
link = 'https://www.pathologyoutlines.com/'
page = requests.get(link)
bs = BeautifulSoup(page.text, 'html.parser')
each_link = bs.find_all('div', class_='float_left')
for j in each_link:
  if index>59:
    break
  new=j.find_all('li')
  name=''
  for n in new:
    new2=n.find_all('a')
    for i in new2:
      index+=1
      name=str(i.text)
      organ_name.insert(cnt, name)
      url = i['href']
      data_generation(cnt,url)
      cnt=cnt+1
```

```python
dictionary = dict(zip(organ_name, organ_content))
print(dictionary.keys())
print(dictionary.values())
```

```python
dataDf = pd.DataFrame()
dataDf['Organ Name'] = organ_name
dataDf['Organ Content'] = organ_content

dataDf
```
